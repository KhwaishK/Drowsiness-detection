## 1. Import Dependencies

```python
from ultralytics import YOLO
from matplotlib import pyplot as plt
import cv2
import os
```

## 2. Load the Trained Model

```python
model = YOLO("runs/detect/train8/weights/best.pt")
```

## 3. Test on Sample Data

#### Sample 1

```python
sample_1 = os.path.join('sample_data', 'sample1.jpg')
```

```python
results = model(sample_1)
```

```python
%matplotlib inline
sample_1 = cv2.cvtColor(results[0].plot(), cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(sample_1)
```

#### Sample 2

```python
sample_2 = os.path.join('sample_data', 'sample2.jpg')
```

```python
results2 = model(sample_2)
```

```python
%matplotlib inline
sample_2= cv2.cvtColor(results2[0].plot(), cv2.COLOR_BGR2RGB)
plt.imshow(sample_2)
```

## 4. Test with Webcam

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():

    ret, frame = cap.read()

    results = model(frame)

    cv2.imshow('Fine Tuned YOLO', results[0].plot())

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
